```python
import numpy as np 
x_train = np.array([[15 , 4, 3,5], [10 ,4,3,1], [33,4,5,2]])
y_train = np.array([50 , 45, 44])
w_ini = np.array([0.001, 0.0001,0.0001,0.0001])
```

```python
def prediction (w,x,b):
    return np.dot(x,w)+b

def cost_fun(pred,y):
    error = 1/(2*len(y))*np.sum((pred - y)**2)
    return error
```

```python
def G_D(x,y,w):
    alpha = 0.001
    b= 0.01
    cost_history = []
    dw_db = []
    iters = 10

    for i in range (iters):
        pred_GD = prediction (w,x,b)

        GD_error = (pred_GD - y)
        dw = 1/(len(y))*np.dot( GD_error ,x)

        db = 1/(len(y))*np.sum(GD_error)

        w = w - alpha*dw
        b = b -alpha *db

        current_cost = cost_fun(pred_GD,y)
        cost_history.append(current_cost)

        dw_db.append({"dw = ":w.copy(), "db = ": b})

    return cost_history,dw_db


cost_history,dw_db = G_D(x_train ,y_train ,w_ini)
print(f"cost history  = {cost_history}")
print(f"dw_db = {dw_db}")
    
    
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression

# Generate a simple dataset
data, target = make_regression(n_samples=100, n_features=1, noise=0.1, random_state=42)

# `data` contains the feature(s)
# `target` contains the output values
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.3,random_state=42)
```

```python
y_test.shape
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
result = model.predict(X_test.reshape(-1, 1))  # Ensure X_test is 2D for prediction
```
